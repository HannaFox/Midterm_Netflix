```python
# creating new empty df called unique
unique_title_df = pd.DataFrame()
```

```python
title = netflix["Title"].unique()
```

```python
unique_title_df["title"] = title
```

```python
unique_title_df.sort_values("title", ascending = True)
```

```python
# merging all relevant IMDb files into IMDb_df
title_tconst_df = pd.read_csv("title.basics.csv")
title_tconst_df.tail()
```

```python
title_tconst_df["titleType"].unique()
```

```python
title_tconst_df=title_tconst_df.replace("tvSeries","TV Show")
```

```python
title_tconst_df=title_tconst_df.replace("movie","Movie")
```

```python
title_tconst_df=title_tconst_df.replace("tvMovie","Movie")
```

```python
title_tconst_df["titleType"].unique()
```

```python
ratings_tconst_df = pd.read_csv("title.ratings.csv")
ratings_tconst_df.tail()
```

```python
IMDb_df = pd.merge(title_tconst_df, ratings_tconst_df, how = "inner", left_on = ["tconst"], right_on = ["tconst"])
IMDb_df.tail()
```

```python
# left merging 642 unique netflix movies with IMDb
netflix_IMDb = pd.merge(unique_title_df,IMDb_df, how = "left", left_on = ["title"], right_on = ["primaryTitle"]) 
netflix_IMDb.tail()
```

```python
sorted_df = netflix_IMDb.sort_values(["startYear"])
```

```python
sorted_df.head()
```

```python
# Drop duplicates except for the last occurrence, keep title that has latest startyear
filtered_df = sorted_df.drop_duplicates(subset='title', keep= "last")
filtered_df.reset_index(inplace=True)
filtered_df.tail()
```

```python
# Problem 2: not every movie in IMDb
filtered_df["title"].nunique()
```

```python
# replacing missing values for averageRating
```

```python
filtered_df["averageRating"].unique()
```

```python
#compute mean
mean_rating = filtered_df["averageRating"].mean()
mean_rating = round(mean_rating,1)
mean_rating
```

```python
filtered_df["averageRating"] = filtered_df["averageRating"].fillna(mean_rating)
```

```python
filtered_df["averageRating"].unique()
```

```python
filtered_df = filtered_df.sort_values(["title"])
filtered_df = filtered_df.reset_index()
```
