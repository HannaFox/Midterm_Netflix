```python
# reading IMDb file
title = pd.read_csv("title.basics.csv")
title.tail()
```

```python
title["titleType"].unique()
```

```python
title=title.replace("tvSeries","TV Show")
```

```python
title=title.replace("movie","Movie")
```

```python
title=title.replace("tvMovie","Movie")
```

```python
title["titleType"].unique()
```

```python
# merging 
left_merged = pd.merge(netflix, title, how = "left", left_on = ["Title","Type"], right_on = ["primaryTitle","titleType"])
left_merged.drop_duplicates()
```

```python
left_merged = left_merged.drop(["As of","Rank","Year to Date Rank","Last Week Rank","Netflix Exclusive", "Netflix Release Date", "Days In Top 10", "Viewership Score", "startYear","endYear"], axis = 1)
left_merged.drop_duplicates()
```
