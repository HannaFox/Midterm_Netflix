```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
netflix = pd.read_excel("netflix_clean_titles_all.xlsx")
```

```python
netflix.tail()
```

```python
netflix["Title"].nunique()
```

```python
# creating new empty df called unique
unique_title_df = pd.DataFrame()
```

```python
title = netflix["Title"].unique()
```

```python
unique_title_df["title"] = title
```

```python
unique_title_df.sort_values("title", ascending = True)
```

```python
# merging all relevant IMDb files
title_tconst_df = pd.read_csv("title.basics.csv")
title_tconst_df.tail()
```

```python
ratings_tconst_df = pd.read_csv("title.ratings.csv")
ratings_tconst_df.tail()
```

```python
IMDb_df = pd.merge(title_tconst_df, ratings_tconst_df, how = "inner", left_on = ["tconst"], right_on = ["tconst"])
```

```python
grouped_imdb= IMDb_df.groupby(['primaryTitle'])['startYear'].max().reset_index()
```

```python
grouped_imdb= IMDb_df.groupby(['primaryTitle'])['startYear'].max().['averageRating'].avg()reset_index()
```

```python
grouped_imdb
```

```python
grouped_imdb.to_excel(r'grouped_IMDb.xlsx', index = False)
```

```python
grouped_imdb.nunique()
```

```python
# left merging 642 unique netflix movies with IMDb
netflix_IMDb = pd.merge(unique_title_df, grouped_imdb, how = "left", left_on = ["title"], right_on = ["primaryTitle"]) 
netflix_IMDb.tail()
```

```python
# Problem 1: not every movie in IMDb
netflix_IMDb["title"].nunique()
```

```python
# Problem 2: a lot of movies have the same title, which is why they are added multiple times to the netflix_IMDb_df
```

```python
# merging 
left_merged = pd.merge(netflix, title, how = "left", left_on = ["Title","Type"], right_on = ["primaryTitle","titleType"])
left_merged.isna().sum()
```

```python
title_df = left_merged[(left_merged["titleType"] == "TV Show") | (left_merged["titleType"] == "Movie") | (left_merged["titleType"] == "nan")]
```
